```python
from sklearn.feature_extraction.text import CountVectorizer

cvect = CountVectorizer(min_df=0.03, max_df=0.95)

df = pd.read_csv(
    path, 
    compression='gzip',
    delimiter="\t"
)

df2transform = df.copy()

df2transform["text"] = df2transform.text.str.replace(
    r'(\s|^)[\d\-\+\=\.\,]+\_NUM', ' numb_mask ', regex=True
)
df2transform["text"] = df2transform["text"].str.replace(r'\b\d+\b', " numb_mask ", regex=True)
bow_arr = cvect.fit_transform(df2transform["text"]).toarray()
```

```python
df_bow = pd.DataFrame(data=bow_arr, columns=cvect.get_feature_names_out())
```
