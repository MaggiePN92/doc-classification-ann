```python
import json

model_scores = {}

# iterate log paths in logs list. 
for log in logs:
    # open each file and parse as json
    with open(f"{folder_with_logs}/{log}", "r") as f:
        modellog = json.loads(f.read())
        # model name, f1 score and training time
        model_name = modellog["model_name"].split("-")[0]
        precision = modellog["clf_rep"]['macro avg']['precision']
        recall = modellog["clf_rep"]['macro avg']['recall']
        f1_score = modellog["clf_rep"]['macro avg']['f1-score']
        accuracy = modellog["accuracy"]

        # put data in dict
        if model_name not in model_scores:
            model_scores[model_name] = {
                "f1_score" : [f1_score],
                "precision" : precision,
                "recall" : recall,
                "f1_score" : f1_score,
                "accuracy" : accuracy
            }
```
